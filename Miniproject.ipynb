```python
X = data1.iloc[:, :41]
y = data1.Class
```

```python
X
```

```python
y
```

```python
X_train, X_test, y_train, y_test =train_test_split(X, y,test_size=0.20, random_state=3)
```

```python
X_train
```

```python
rbf = svm.SVC(kernel='rbf', gamma=0.5, C=0.1).fit(X_train, y_train)
poly = svm.SVC(kernel='poly', degree=3, C=1).fit(X_train, y_train)
lin = svm.SVC(kernel='linear').fit(X_train, y_train)
```

```python
poly_pred = poly.predict(X_test)
rbf_pred = rbf.predict(X_test)
lin_pred = lin.predict(X_test)
```

```python
rbf.score(X_test,y_test)
```

```python
poly_accuracy = accuracy_score(y_test, poly_pred)
rbf_accuracy = accuracy_score(y_test, rbf_pred)
lin_accuracy = accuracy_score(y_test, lin_pred)


print('Accuracy (Polynomial Kernel): ', "%.2f" % (poly_accuracy*100))
print('Accuracy (rbf Kernel): ', "%.2f" % (rbf_accuracy*100))
print('Accuracy (liner Kernel): ', "%.2f" % (lin_accuracy*100))
```

```python
cv_score_svm = cross_val_score(svm.SVC(kernel='rbf'), X, y, cv=5)

print(cv_score_svm)

mean_accuracy_svm = sum(cv_score_svm)/len(cv_score_svm)

mean_accuracy_svm = mean_accuracy_svm*100

mean_accuracy_svm = round(mean_accuracy_svm, 2)

print(mean_accuracy_svm)
```
